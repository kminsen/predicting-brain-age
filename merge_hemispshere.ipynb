```python
# gaussl_df = pd.read_csv('D:\BrainAge Dataset\Train\Train\lh.GausCurv - 1600.csv')
# gaussr_df = pd.read_csv('D:\BrainAge Dataset\Train\Train\\rh.GausCurv- 1600.csv')

gaussl_df = dataframe.filter(regex = 'lh.GausCurv_')
gaussr_df = dataframe.filter(regex = 'rh.GausCurv_')

gauss_df = pd.DataFrame()
for id in brain_regions:
    gauss_df[id] = gaussl_df['lh.GausCurv_' + id] + gaussr_df['rh.GausCurv_' + id]

gauss_df =gauss_df.add_prefix('GausCurv_',axis = 1)
```

```python
# meanl_df = pd.read_csv('D:\BrainAge Dataset\Train\Train\lh.MeanCurv - 1600.csv')
# meanr_df = pd.read_csv('D:\BrainAge Dataset\Train\Train\\rh.MeanCurv- 1600.csv')

meanl_df = dataframe.filter(regex = 'lh.MeanCurv_')
meanr_df = dataframe.filter(regex = 'rh.MeanCurv_')

mean_df = pd.DataFrame()
for id in brain_regions:
    mean_df[id] = meanl_df['lh.MeanCurv_'+id] + meanr_df['rh.MeanCurv_'+id]

mean_df = mean_df.add_prefix('MeanCurv_',axis = 1)
```

```python
# grayvoll_df = pd.read_csv('D:\BrainAge Dataset\Train\Train\lh.GrayVol - 1600.csv')
# grayvolr_df = pd.read_csv('D:\BrainAge Dataset\Train\Train\\rh.GrayVol- 1600.csv')

grayvoll_df = dataframe.filter(regex = 'lh.GrayVol_')
grayvolr_df = dataframe.filter(regex = 'rh.GrayVol_')

grayvol_df = pd.DataFrame()
for id in brain_regions:
    grayvol_df[id] = grayvoll_df['lh.GrayVol_'+id] + grayvolr_df['rh.GrayVol_'+id]

grayvol_df = grayvol_df.add_prefix('GrayVol_')
```

```python
# thickavgl_df = pd.read_csv('D:\BrainAge Dataset\Train\Train\lh.ThickAvg - 1600.csv')
# thickavgr_df = pd.read_csv('D:\BrainAge Dataset\Train\Train\\rh.ThickAvg- 1600.csv')

thickavgl_df = dataframe.filter(regex = 'lh.ThickAvg_')
thickavgr_df = dataframe.filter(regex = 'rh.ThickAvg_')

thickavg_df = pd.DataFrame()
for id in brain_regions:
    thickavg_df[id] = thickavgl_df['lh.ThickAvg_'+id] + thickavgr_df['rh.ThickAvg_'+id]

thickavg_df =thickavg_df.add_prefix('ThickAvg_',axis = 1)
```

```python
# surfacel_df = pd.read_csv('D:\BrainAge Dataset\Train\Train\lh.SurfArea - 1600.csv')
# surfacer_df = pd.read_csv('D:\BrainAge Dataset\Train\Train\\rh.SurfArea - 1600.csv')

surfacel_df = dataframe.filter(regex = 'lh.SurfArea_')
surfacer_df = dataframe.filter(regex = 'rh.SurfArea_')

surface_df = pd.DataFrame()
for id in brain_regions:
    surface_df[id] = surfacel_df['lh.SurfArea_' + id] + surfacer_df['rh.SurfArea_' + id]

surface_df = surface_df.add_prefix('SurfArea_',axis = 1)
```

```python
# aseg = pd.read_csv('D:\BrainAge Dataset\Train\Train\\aseg - 1600.csv')
aseg_l = dataframe.filter(regex = 'aseg_Left-')
aseg_l.columns = aseg_l.columns.str.replace('aseg_Left-','')
aseg_r = dataframe.filter(regex = 'aseg_Right-')
aseg_r.columns = aseg_r.columns.str.replace('aseg_Right-','')

aseg_regions = aseg_l.columns

aseg_df = pd.DataFrame()
for region in aseg_regions:
    aseg_df[region] = aseg_l[region] + aseg_r[region]

for region in ['Brain-Stem','CC_Anterior','CC_Mid_Anterior','CC_Central','CC_Mid_Posterior','CC_Posterior','Optic-Chiasm']:
    aseg_df[region] = dataframe['aseg_'+region]

aseg_df = aseg_df.add_prefix('aseg_',axis = 1)
```

```python
# wmparc = pd.read_csv('D:\BrainAge Dataset\Train\Train\\wmparc - 1600.csv')
# print(wmparc.columns.shape)

wmparc_l = dataframe.filter(regex = 'wmparc_wm-lh-')
wmparc_r = dataframe.filter(regex = 'wmparc_wm-rh-')
print(wmparc_l.columns.shape)

remaining = ['Left-UnsegmentedWhiteMatter','Right-UnsegmentedWhiteMatter']

wmparc_l.columns = wmparc_l.columns.str.replace('wmparc_wm-lh-','')
wmparc_r.columns = wmparc_r.columns.str.replace('wmparc_wm-rh-','')

wmparc_regions = wmparc_l.columns

wmparc_df = pd.DataFrame()
for region in wmparc_regions:
    wmparc_df[region] = wmparc_l[region] + wmparc_r[region]

wmparc_df['UnsegmentedWhiteMatter'] = dataframe['wmparc_Left-UnsegmentedWhiteMatter'] + dataframe['wmparc_Right-UnsegmentedWhiteMatter']

wmparc_df = wmparc_df.add_prefix('wmparc_',axis = 1)

```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Data/merged_hemisphere.csv')
original_columns = df.columns
```

```python
num_bins = 32
df['binned_age'] = pd.qcut(df['age'], q=num_bins, labels=[i for i in range(num_bins)])
```

```python
df['binned_age'].unique()
```

```python
grouped_means = df.groupby('binned_age').mean().reset_index()
grouped_medians = df.groupby('binned_age').median().reset_index()
grouped_stdev = df.groupby('binned_age').std().reset_index()

# Merge the mean values back to the original DataFrame
df = df.merge(grouped_means, on='binned_age', suffixes=('', '_groupmean'))
df = df.merge(grouped_medians, on='binned_age', suffixes=('', '_groupmedian'))
df = df.merge(grouped_stdev, on='binned_age', suffixes=('', '_groupstd'))
```

```python
df.drop(columns=original_columns,inplace = True)
df
```

```python
test = pd.read_csv('Data/binned_on_age_original_quartile8.csv')
a = set(df.columns.unique())
b = set(test.columns.unique())

discard = a-b
print((a-b))
```

```python
df.drop(columns = list(discard),inplace = True)
```
